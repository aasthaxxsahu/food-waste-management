```python
#EDA 1: Provider Type Distribution
plt.figure(figsize=(8,5))
sns.countplot(data=providers, x='Type')

plt.title('Provider Type Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
#EDA 4: Meal Type Distribution
plt.figure(figsize=(8,5))
sns.countplot(data=food, x='Meal_Type')

plt.title('Meal Type Distribution')
plt.show()
```

```python
# Bivariate Analysis
#EDA 5. City vs Food Listings
food['Location'].value_counts().plot(kind='bar')
```

```python
#EDA 6. Provider Type vs Quantity
food.groupby('Provider_Type')['Quantity'].sum().plot(kind='bar')
```

```python
#EDA 7. Food Type vs Quantity
food.groupby('Food_Type')['Quantity'].sum().plot(kind='bar')
```

```python
#EDA 8. Meal Type vs Quantity
food.groupby('Meal_Type')['Quantity'].sum().plot(kind='bar')
```

```python
# Multivariate Analysis

#First merge datasets:

merged = food.merge(
    providers[['Provider_ID','Type','City']],
    on='Provider_ID',
    how='left'
)
```

```python
#EDA 9. City + Provider Type + Quantity
pivot = merged.pivot_table(
    values='Quantity',
    index='City',
    columns='Type',
    aggfunc='sum'
)

sns.heatmap(pivot, cmap='YlGnBu')
plt.show()
```

```python
#EDA 10. Food Type + Meal Type + Quantity
pivot = food.pivot_table(
    values='Quantity',
    index='Food_Type',
    columns='Meal_Type',
    aggfunc='sum'
)

sns.heatmap(pivot, cmap='coolwarm')
plt.show()
```

```python
#EDA 11. Provider + Claims + Quantity
provider_claims = food.merge(
    claims,on='Food_ID'
).groupby('Provider_ID').agg({
    'Claim_ID':'count',
    'Quantity':'sum'
})

provider_claims.head(10).plot(kind='bar')
```

```python
#EDA12. Receiver + Claims + Quantity
receiver_claims = claims.merge(
    food,on='Food_ID'
).groupby('Receiver_ID').agg({
    'Claim_ID':'count',
    'Quantity':'sum'
})

receiver_claims.head(10).plot(kind='bar')
```

# Business Insights

### Food Availability
The analysis shows that New Carol and South Kathryn have the highest number of food listings. This indicates that these cities have more food available for donation compared to other locations.

### Food Waste
Among all meal categories, Breakfast recorded the highest number of claims. This suggests that breakfast items are in high demand and form a significant portion of food distribution activities.

### Provider Analysis
Barry Group was identified as the top contributing provider. It donated the highest quantity of food and also recorded the highest number of successful food claims.

### Receiver Analysis
William Frederick, Matthew Webb, Scott Hunter, and Anthony Garcia were among the most active receivers, each making multiple food claims through the platform.

### Claims Analysis
The claim status analysis showed that 33.9% of claims were completed, while the remaining claims were either pending or cancelled. This highlights opportunities to improve the claim fulfillment process.

### Demand Analysis
Cities such as Lake Rebeccaton, Emilymouth, South Laurachester, Andersonview, and East Michael showed the highest demand based on the number of completed food claims.

# Business Recommendations

### Recommendation 1
Cities with a high number of food listings should collaborate more closely with NGOs and community organizations to ensure that surplus food reaches people in need efficiently.

### Recommendation 2
Top contributing providers should be recognized and encouraged to continue their participation, as they play an important role in reducing food wastage.

### Recommendation 3
An automated notification system should be implemented to alert providers before food items reach their expiry date, helping reduce avoidable food waste.

### Recommendation 4
Food collection and distribution efforts should be increased in cities with high demand to ensure that available food reaches receivers more effectively.

### Recommendation 5
Efforts should be made to reduce cancelled and pending claims by improving communication and coordination between providers and receivers.
